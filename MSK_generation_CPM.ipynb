```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from vlfpy import *
import matplotlib as mpl
mpl.rcParams['mathtext.default'] = 'regular'
```

```python

data = np.array([1, 0, 1, 1, 0, 0, 0, 1])
T = 0.1
fc = 1/T*1.25
fs = 1000

s, t, phase = generate_msk_data(data, T, fc, fs)

fig, axs = plt.subplots(3, 1, figsize=(9, 4), sharex=True)

data_plot = np.zeros_like(t)
for n in range(len(data)):
    t1 = int(n*T*fs)
    t2 = int((n+1)*T*fs)
    data_plot[t1:t2] = data[n]*2-1

axs[0].plot(t/T, data_plot)


for i in range(1, int(len(data))):
    axs[0].axvline(x=i*T*10, color='black', linestyle='--', alpha=0.5)
for i in range(len(data)):
    axs[0].text( i*T*10+0.25,-0.2, '$a_'  + str(i) + '$ = ' + str(data[i]))
axs[0].set_title('Data')

phase_plot = phase[phase < 2e-2] = np.nan 
axs[1].plot(t/T, phase/np.pi)
for i in range(1, int(len(data))):
    axs[1].axvline(x=i*T*10, color='black', linestyle='--', alpha=0.5)
axs[1].set_title('Phase')
axs[1].set_yticks([0, 1, 2])
axs[1].set_yticklabels(['0', '$\pi$', '$2\pi$'])
axs[1].grid()

axs[2].plot(t/T, s)
for i in range(1, int(len(data))):
    axs[2].axvline(x=i*T*10, color='black', linestyle='--', alpha=0.5)
axs[2].set_title('MSK Signal')
axs[2].set_xlabel('t')
x_labels = ['0', 'T', '2T', '3T', '4T', '5T', '6T', '7T', '8T']

axs[2].set_xticks(np.arange(0, len(data)+1, 1))
axs[2].set_xticklabels(x_labels)
plt.tight_layout()


#plt.savefig('figures/theory/msk_cpm.pdf')

```

# introducing offset

```python

data = np.array([1, 0, 1, 1, 0, 0, 0, 1])
T = 0.1
fc = 1/T*1.25
fs = 1000
s1, t_1, phase1 = generate_msk_data(data, T, fc, fs, offset=0.2*T)

fig, axs = plt.subplots(3, 1, figsize=(9, 4), sharex=True)

data_plot = np.zeros_like(t_1)
for n in range(len(data)):
    t1 = int(n*T*fs)
    t2 = int((n+1)*T*fs)
    data_plot[t1:t2] = data[n]

axs[0].plot(t_1/T, data_plot)
for i in range(1, int(len(data))):
    axs[0].axvline(x=i*T*10, color='black', linestyle='--')
for i in range(len(data)):
    axs[0].text(i*T*10+0.45,0.5, data[i])
axs[0].set_title('Data')

phase_plot1 = phase1[phase1 < 2e-2] = np.nan 
axs[1].plot(t_1/T, phase1/np.pi)
for i in range(1, int(len(data))):
    axs[1].axvline(x=i*T*10, color='black', linestyle='--')
axs[1].set_title('Phase')

axs[2].plot(t_1/T, s1)
for i in range(1, int(len(data))):
    axs[2].axvline(x=i*T*10, color='black', linestyle='--')
axs[2].set_title('MSK Signal')

plt.tight_layout()
```
